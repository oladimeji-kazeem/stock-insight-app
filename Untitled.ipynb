```python
def clean_text(text):
    text = re.sub(r"http\S+", "", text)  # Remove URLs
    text = re.sub(r"[^A-Za-z0-9\s]", "", text)  # Remove special characters
    text = re.sub(r"\s+", " ", text)  # Replace multiple spaces with one
    return text.strip().lower()
```

```python
def fetch_news(api_key, company="Apple", language="en", max_articles=10):
    url = (
        f"https://newsapi.org/v2/everything?"
        f"q={company}&"
        f"language={language}&"
        f"pageSize={max_articles}&"
        f"sortBy=publishedAt&"
        f"apiKey={api_key}"
    )
    response = requests.get(url)
    data = response.json()

    articles = [{
        "date": article["publishedAt"][:10],
        "headline": article["title"],
        "source": article["source"]["name"]
    } for article in data.get("articles", [])]

    return pd.DataFrame(articles)

# Replace YOUR_API_KEY with your real key
api_key = "0ecb64bff77142dda40e88df0e7d362f"
news_df = fetch_news(api_key, company="Apple", max_articles=10)
news_df["cleaned_headline"] = news_df["headline"].apply(clean_text)
print(news_df.head())
```

```python
# Load FinBERT model and tokenizer
model_name = "yiyanghkust/finbert-tone"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
finbert = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Apply to headlines or tweets
news_df["sentiment_result"] = news_df["cleaned_headline"].apply(finbert)
news_df["sentiment"] = news_df["sentiment_result"].apply(lambda x: x[0]["label"])
```
